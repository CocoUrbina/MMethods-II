las $n$ raíces $n$-ésimas del número complejo $z$.

\begin{align}
S &= \sum_{k=0}^{n-1} \sqrt[n]{r}e ^{i(\theta + 2 \pi k) / n} \\

\begin{align}

\end{align}

\end{align}
